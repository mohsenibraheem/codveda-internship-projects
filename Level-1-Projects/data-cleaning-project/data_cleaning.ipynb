## Cleaning the data

### 1. The "Transaction ID" column has no null values, we leave it.

### 2. Fill with "Unknown" for "Item" column 

```python
# Make the data consistent
cafe_sales_df["Item"] = cafe_sales_df["Item"].replace(["UNKNOWN", "Unknown", "ERROR", "Unknown"])
cafe_sales_df["Item"] = cafe_sales_df["Item"].fillna("Unknownn") 
```

```python
cafe_sales_df["Item"].value_counts(dropna=False)
```

### 3. For "Quantity" column, fill with median for better accurate guess and making balanced data

```python
# count unique values
cafe_sales_df["Quantity"].value_counts(dropna=False)
```

```python
# convert characters to NaN (Null values)
cafe_sales_df["Quantity"] = pd.to_numeric(cafe_sales_df["Quantity"], errors = "coerce")
```

```python
# Fill with median of the data
cafe_sales_df["Quantity"] = cafe_sales_df["Quantity"].fillna(cafe_sales_df["Quantity"].median())
```

### 4. Fill "Price per quantity" column with the it's median

```python
# convert records to null values
cafe_sales_df["Price Per Unit"] = pd.to_numeric(cafe_sales_df["Price Per Unit"], errors = "coerce")
```

```python
# Impute the missing reocords with the column's median number
cafe_sales_df["Price Per Unit"]= cafe_sales_df["Price Per Unit"].fillna(cafe_sales_df["Price Per Unit"].median())
```

### 5. Recalculate total spent for better accuracy.

```python
# Total = Price per unit X Quantity
cafe_sales_df["Total Spent"] = cafe_sales_df["Price Per Unit"] * cafe_sales_df["Quantity"]
```

### 6. Fix  "Payment Method" columns

```python
# fix bad missing 
cafe_sales_df["Payment Method"] = cafe_sales_df["Payment Method"].replace(["ERROR", "Unknown", "UNKNOWN", "Unknown"])
# Now fill all missing values with "Unknown"
cafe_sales_df["Payment Method"] = cafe_sales_df["Payment Method"].fillna("Unknown")
```

### 7. Clean the "Location" column

```python
# count the values of the column
cafe_sales_df["Location"].value_counts(dropna = False)
```

```python
cafe_sales_df["Location"] = cafe_sales_df["Location"].replace(["ERROR", "Unknown"])
# make the column consistence
cafe_sales_df["Location"] = cafe_sales_df["Location"].replace(["UNKNOWN", "Unknown"])
```

```python
# Fill the null records with 'Unknown'
cafe_sales_df["Location"] = cafe_sales_df["Location"].fillna("Unknown")
```

```python
cafe_sales_df["Location"].head()
```

```python
cafe_sales_df["Location"].value_counts(dropna = False)
```

### 8. Clean "Transaction Date" column

```python
cafe_sales_df["Transaction Date"].value_counts(dropna = False)
```

```python
cafe_sales_df["Transaction Date"] = pd.to_datetime(cafe_sales_df["Transaction Date"], errors = "coerce")
```

#### Since the date column is useful and has small size of empty records, 
#### We leave it like that for safer and honest analysis.
